```python

import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import figure
from ipywidgets import interact
import matplotlib.gridspec as gridspec

plt.rc('font', size=15)         
plt.rc('axes', labelsize=20)

def muplus(beta):
    return 0.25*(1+np.sqrt(1+4/(beta**2)))*(1+beta/np.sqrt(beta**2+4))

def muminus(beta):
    return 0.25*(1-np.sqrt(1+4/(beta**2)))*(1-beta/np.sqrt(beta**2+4))

def thetaplus(beta):
    return 0.5*(beta+np.sqrt(beta**2+4))

def thetaminus(beta):
    return 0.5*(beta-np.sqrt(beta**2+4))


def graphs(b):

    betas = np.linspace(0.1,10,1000)

    mup = muplus(betas*b)
    mum = muminus(betas*b)
    
    plt.figure(1)
    gs = gridspec.GridSpec(2, 1,height_ratios=[1,1])

    ax1=plt.subplot(gs[0])
    line1 = plt.plot(betas,thetaplus(betas*b),'r',label=r'$\mu+$')
    line2 = plt.plot(betas,thetaminus(betas*b),'b',label=r'$\mu-$')
    plt.xticks([])
    plt.legend()
    plt.ylabel(r'$\theta$')

    ax2=plt.subplot(gs[1])
    line1=plt.plot(betas,mup,'r',label=r'$\mu+$')
    line2=plt.plot(betas,mum,'b',label=r'$\mu-$')
    plt.xlabel(r'$\frac{\beta}{\theta_E}$')
    plt.ylabel(r'$\mu$')
    plt.legend()
    plt.show()
    
graphs(1)
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import figure
from ipywidgets import interact
from scipy.interpolate import Rbf
from scipy.interpolate import griddata
import matplotlib.gridspec as gridspec
from scipy.interpolate import LinearNDInterpolator

plt.rcParams['figure.figsize'] = [10, 10]

#source brightness distribution profiles
def rad(x,y,s):
    return 1.5*s/(np.sqrt(x**2+y**2)+0.1)

def radsqr(x,y,s):
    return s/(x**2+y**2+0.1)

def exp(x,y,s):
    return 15*np.exp(-(s*np.sqrt(x**2+y**2))+0.1)

def twopoints(x,y,s):
    return 1.5/(np.sqrt((x-s)**2+(y)**2))+1.5/(np.sqrt((x+s)**2+(y)**2)+0.01)

def multipoint(x,y,nn,centt_x,centt_y,peak):
    
    bri = np.zeros(nn)
    
    for i in range(nn):
        bri[i] = peak[i]/((x-centt_x[i])**2+(y-centt_y[i])**2+0.01)
    
    bb = np.sum(bri)
                       
    return bb

def point(x,y,s):
    if np.abs(x)<0.1*s and np.abs(y)<0.1*s:
        return 10
    else:
        return 0

#the random positions of all the points in multipoint
cent_x = np.random.uniform(low=-4.0, high=4.0, size=100)
cent_y = np.random.uniform(low=-2.0, high=2.0, size=100)
peaks = np.random.uniform(low=0.0, high=0.1, size=100)


#x and y coordinates for plotting the Einstein radius
xcircle = np.linspace(-1,1,100)
ycircle = np.sqrt(1-xcircle**2)
ycircle2 = -np.sqrt(1-xcircle**2)

#lensing
def thetaSIS(beta):
    if beta > 1:
        return beta - 1, 0
    else:
        return beta - 1, beta + 1

def thetaPM(beta):
    return 0.5*(beta+np.sqrt(beta**2+4)), 0.5*(beta-np.sqrt(beta**2+4))

#plotting an interactive graph
def graph(fineness=200,profile='point',lens='point mass',number_of_lenses=1,s_parameter=2,centre_x=0.0,centre_y=0.0, lens2xcent = 0.0, lens2ycent = 0.0):

    if profile == 'rad':
        prof = rad
    elif profile == 'radsqr':
        prof = radsqr
    elif profile == 'exp':
        prof = exp
    elif profile =='two points':
        prof = twopoints
    elif profile=='multiple points':
        prof = multipoint
    elif profile=='point':
        prof = point
        
    if lens == 'point mass':
        lenss = thetaPM
        lensbright = point
        
    elif lens == 'SIS':
        lenss = thetaSIS
        lensbright = radsqr
    
    #max X and Y values out to which we plot the brightness distribution 
    maxx = 6
    
    #the x and y coordinates of all points in the source
    x = np.linspace(-maxx,maxx,fineness)
    y = np.linspace(-maxx,maxx,fineness)
    
    #we need to turn this into a grid, and so
    X,Y  = np.meshgrid(x,y)
    
    #creating an array to hold the values of brightness
    brightnesses = np.zeros(X.shape)
    lens_flux = np.zeros(X.shape)
    
    #calculating the brightnesses for the choice of profile
    if profile == 'multiple points':
    
        n = int(round(10*s_parameter))

        for i in range(len(x)):
            for j in range(len(y)):
                brightnesses[i][j] = prof(X[i][j],Y[i][j],n,cent_x,cent_y,peaks)
                lens_flux[i][j] = lensbright(X[i][j],Y[i][j],0.5)
                
                #shifting the centre of the source
                X[i][j] += centre_x
                Y[i][j] += centre_y
    
    else:
        for i in range(len(x)):
            for j in range(len(y)):
                brightnesses[i][j] = prof(X[i][j],Y[i][j],0.5*s_parameter)
                lens_flux[i][j] = lensbright(X[i][j],Y[i][j],0.5)                
                #shifting the centre of the source
                X[i][j] += centre_x
                Y[i][j] += centre_y
    
    
    #creating arrays to hold the new coordinates of each brightness point
    thetapx = np.zeros(X.shape)
    thetapy = np.zeros(Y.shape)

    thetamx = np.zeros(X.shape)
    thetamy = np.zeros(Y.shape)

    thetapx2 = np.zeros(X.shape)
    thetapy2 = np.zeros(Y.shape)

    thetamx2 = np.zeros(X.shape)
    thetamy2 = np.zeros(Y.shape)

    relpx = np.zeros(X.shape)
    relpy = np.zeros(Y.shape)

    relmx = np.zeros(X.shape)
    relmy = np.zeros(Y.shape)
    
    #transforming the coordinates into the lens plane
    for i in range(len(x)):
        for j in range(len(y)):
            #converting to polar coordinates
            phi = math.atan2(Y[i][j],X[i][j]) 
            beta = np.sqrt(X[i][j]**2+Y[i][j]**2) #point mass lensing depends only on the distance from the origin
        
            rp,rm = lenss(beta) #since point mass lensing occurs on a line, can apply this scalar equation
        
            #converting back into cartesian coordinates
            thetapx[i][j] = rp*np.cos(phi)
            thetapy[i][j] = rp*np.sin(phi)

            thetamx[i][j] = rm*np.cos(phi)
            thetamy[i][j] = rm*np.sin(phi)
            
    if number_of_lenses==2:
        for i in range(len(x)):
            for j in range(len(y)):
                
                #finding the coordinates relative the the new origin defined by the 2nd lens
                relpx[i][j] = thetapx[i][j] - lens2xcent
                relpy[i][j] = thetapy[i][j] - lens2ycent
                
                relmx[i][j] = thetamx[i][j] - lens2xcent
                relmy[i][j] = thetamy[i][j] - lens2ycent
                
                #converting to polar coordinates
                phi = math.atan2(relpy[i][j],relpx[i][j]) 
                beta = np.sqrt(relpx[i][j]**2+relpy[i][j]**2) #point mass lensing depends only on the distance from the origin
        
                rp,rm = lenss(beta) #since point mass lensing occurs on a line, can apply this scalar equation
        
                #converting back into cartesian coordinates
                thetapx2[i][j] = rp*np.cos(phi)
                thetapy2[i][j] = rp*np.sin(phi)

                thetamx2[i][j] = rm*np.cos(phi)
                thetamy2[i][j] = rm*np.sin(phi)        
    
    #sets the colour gradients
    levlist = np.arange(0,10,0.2)
    
    if number_of_lenses == 1:
        
        plt.figure(1)
        gs = gridspec.GridSpec(2, 1,height_ratios=[1,1])

        ax1=plt.subplot(gs[0])
        plt.title('Source Plane')
        line1 = plt.contourf(X,Y,brightnesses,levels=levlist,cmap=plt.get_cmap('hot'))
        #line5 = plt.contourf(X,Y,lens_flux,levels=levlist,cmap=plt.get_cmap('hot'))
        line2 = plt.plot(0,0,'bx')
        line3 = plt.plot(xcircle,ycircle,'b--')
        line4 = plt.plot(xcircle,ycircle2,'b--')
        plt.xlim(-3,3)
        plt.ylim(-1.5,1.5)
        plt.gca().set_aspect('equal')
        plt.colorbar()
    
        #X_edges, Y_edges = np.meshgrid(edges, edges)
    
        ax2=plt.subplot(gs[1])
        plt.title('Image Plane')
        line1 = plt.contourf(thetapx,thetapy,brightnesses,levels=levlist, cmap=plt.get_cmap('hot'))
        line2 = plt.contourf(thetamx,thetamy,brightnesses,levels=levlist, cmap=plt.get_cmap('hot'))
        line3 = plt.plot(0,0,'bx')
        line4 = plt.plot(xcircle,ycircle,'b--')
        line5 = plt.plot(xcircle,ycircle2,'b--')
        plt.xlim(-3,3)
        plt.ylim(-1.5,1.5)
        plt.gca().set_aspect('equal')
        plt.colorbar()

    elif number_of_lenses == 2:
        
        plt.figure(1)
        gs = gridspec.GridSpec(3, 1,height_ratios=[1,1,1])

        ax1=plt.subplot(gs[0])
        plt.title('Source Plane')
        line1 = plt.contourf(X,Y,brightnesses,levels=levlist,cmap=plt.get_cmap('hot'))
        #line5 = plt.contourf(X,Y,lens_flux,levels=levlist,cmap=plt.get_cmap('hot'))
        line2 = plt.plot(0,0,'bx')
        line3 = plt.plot(lens2xcent,lens2ycent,'bx')
        #line3 = plt.plot(xcircle,ycircle,'b--')
        #line4 = plt.plot(xcircle,ycircle2,'b--')
        plt.xlim(-3,3)
        plt.ylim(-1.5,1.5)
        plt.gca().set_aspect('equal')
        plt.colorbar()
    
        #X_edges, Y_edges = np.meshgrid(edges, edges)
    
        ax2=plt.subplot(gs[1])
        plt.title('1st Lens Image Plane')
        line1 = plt.contourf(thetapx,thetapy,brightnesses,levels=levlist, cmap=plt.get_cmap('hot'))
        line2 = plt.contourf(thetamx,thetamy,brightnesses,levels=levlist, cmap=plt.get_cmap('hot'))
        line3 = plt.plot(0,0,'bx')
        #line4 = plt.plot(xcircle,ycircle,'b--')
        #line5 = plt.plot(xcircle,ycircle2,'b--')
        plt.xlim(-3,3)
        plt.ylim(-1.5,1.5)
        plt.gca().set_aspect('equal')
        plt.colorbar()
        
        ax3=plt.subplot(gs[2])
        plt.title('2nd Lens Image Plane')
        line1 = plt.contourf(thetapx2,thetapy2,brightnesses,levels=levlist, cmap=plt.get_cmap('hot'))
        line2 = plt.contourf(thetamx2,thetamy2,brightnesses,levels=levlist, cmap=plt.get_cmap('hot'))
        #line3 = plt.plot(0,0,'bx')
        line4 = plt.plot(lens2xcent,lens2ycent,'bx')
        #line5 = plt.plot(xcircle,ycircle,'b--')
        #line6 = plt.plot(xcircle,ycircle2,'b--')
        plt.xlim(-3,3)
        plt.ylim(-1.5,1.5)
        plt.gca().set_aspect('equal')
        plt.colorbar()
            
interact(graph, fineness = [10,20,50,70,100,200,500,1000],profile=['rad','radsqr','exp','two points','multiple points','point'],lens=['point mass','SIS'],number_of_lenses=[1,2], s_parameter = (0,5,0.2), centre_x=(-3,3,0.01),centre_y=(-3,3,0.01),lens2xcent = (-2,2,0.02), lens2ycent = (-2,2,0.02))
plt.show()
```
