# Statistics

In this notebook, I want to examine the usefulness of time delay data when modelling gravitationally lensed systems. To this end, I will consider three models:

1. A strongly-lensed source galaxy with no quasar or time delays
2. A strongly-lensed source galaxy + quasar with 4 point images, but no time delays
3. A strongly-lensed source galaxy + quasar with 4 point images, with time delays for these images

### The things we don't change

```python
## Cosmology

#these are typical values used in a variety of places
H_0 = 70
omega_mm = 0.3
cosmo = FlatLambdaCDM(H0=70, Om0=0.3, Ob0=0.)

## Data

#match the WFC on Hubble 
deltaPix = 0.05

#see e.g. Birrer 2016 Mass sheet degeneracy and time-delay cosmography
psf_type = 'GAUSSIAN'  # 'GAUSSIAN', 'PIXEL', 'NONE'

#typical Hubble exposure time (e.g. COSMOS overview 2006)
exp_time = 1600.

#typical Hubble FWHM in pixels (see https://hst-docs.stsci.edu/)
fwhm = 0.15

#supersampling involves scaling the image up and then back down, to smooth out jagged edges caused by the pixels
#I think that setting the supersampling factor to 1 and convolution to false effectively means that this step
#isn't done
kwargs_numerics = {'supersampling_factor': 1, 'supersampling_convolution': False}

# Choice of source type
source_model_list = ['SERSIC_ELLIPSE']
source_model_class = LightModel(source_model_list)

point_source_list = ['LENSED_POSITION']
point_source_class = PointSource(point_source_type_list=point_source_list, fixed_magnification_list=[False])
```

### upper and lower bounds for randomly-sampled parameters

```python
###cosmology

#from "population of galaxy galaxy strong lenses"
z_source_min = 0.1
z_source_max = 5

z_lens_frac_min = 0.1
z_lens_frac_max = 0.95

#to avoid having a lens with a higher redshift than the source,
#we determine the lens redshift by randomly sampling a fraction of the source redshift

##data

snr_min = 10**2
snr_max = 10**4

##lens model

#lens_models = ['EPL', 'NIE', 'SPEMD', 'SPEP']
lens_models = ['EPL']

lens_ellipticity_min = -0.1
lens_ellipticity_max = 0.1

#galaxy mass (in kg)
M_min = constants.M_sun*(10**10)
M_max = constants.M_sun*(10**11) 

#velocity dispersion (in m/s)
v_disp_mean = 210*(10**3)
v_disp_sigma = 54*(10**3)

theta_E_min = 1     #0.5 might be better

gamma_min = 1.5
gamma_max = 2.5

LOS_gamma_min = -0.1
LOS_gamma_max = 0.1

##source model

source_ellipticity_min = -0.1
source_ellipticity_max = 0.1

source_pos_min = -0.1
source_pos_max = 0.1

source_amp_min = 1000
source_amp_max = 1000

#radius (in pc)
R_min = 10000
R_max = 10000

n_sersic_min = 0.5
n_sersic_max = 10

##point source model

#the factor by which the source amplitude is multiplied by to give the point source amplitude
amp_factor_min = 10
amp_factor_max = 50

#time delays
dt_sigma_factor_min = 0.01
dt_sigma_factor_max = 0.05

n_models = 30
```

### choosing values

```python
###cosmology

z_source = []
z_lens = []

##data

background_rms = []
numPix = []

##lens model

dominant_model = []

lens_ellipticity = []
theta_E = []
gamma = []

gamma_od = []
gamma_os = []
gamma_ds = []

gamma_los = []

##source model
                    
source_ellipticity = []

source_amp = []
    
R_sersic = []

n_sersic = []
    
##point source model
    
amp_factor = []


for i in range(n_models):
    
    ###cosmology
    
    z_s = random.uniform(z_source_min, z_source_max)
    M_lens = random.uniform(M_min, M_max)
    v_disp = np.random.normal(v_disp_mean,v_disp_sigma)
    
    z_source.append(z_s)
    
    bad_values = True
    
    while bad_values:
        
        z_l = random.uniform(z_lens_frac_min, z_lens_frac_max)*z_s
        
        lensCosmo = LensCosmo(z_lens=z_l, z_source=z_s, cosmo=cosmo)
        d_ls = lensCosmo.dds*constants.Mpc
        d_l = lensCosmo.dd*constants.Mpc
        d_s = lensCosmo.ds*constants.Mpc
        
        tE = (1/constants.arcsec)*np.sqrt(4*constants.G*M_lens*d_ls/(d_l*d_s*constants.c**2))
        
        #tE = (1/constants.arcsec)*4*np.pi*(v_disp**2/constants.c**2)*(d_ls/d_s)

        if tE > theta_E_min:
            bad_values = False
            if tE > 5:
                print('')
                print('theta_E = ', tE)
                print('M_lens = ', M_lens)
                print('z_l = ', z_l)
                print('z_s = ', z_s)
        
    theta_E.append(tE)        
    z_lens.append(z_l)


    numPix.append(int(round(2*tE)/0.025))    
    
    ##data

    snr = random.uniform(snr_min, snr_max)
    
    background = 1/snr
    
    background_rms.append(background)

    ##lens model
    
    dominant_model.append(random.choice(lens_models))

    lens_ellipticity.append([random.uniform(lens_ellipticity_min, lens_ellipticity_max), random.uniform(lens_ellipticity_min, lens_ellipticity_max)])

    gamma.append(random.uniform(gamma_min,gamma_max))

    gamma_od.append(np.array([random.uniform(LOS_gamma_min, LOS_gamma_max), random.uniform(LOS_gamma_min, LOS_gamma_max)]))
    gamma_os.append(np.array([random.uniform(LOS_gamma_min, LOS_gamma_max), random.uniform(LOS_gamma_min, LOS_gamma_max)]))
    gamma_ds.append(np.array([random.uniform(LOS_gamma_min, LOS_gamma_max), random.uniform(LOS_gamma_min, LOS_gamma_max)]))

    gl = gamma_os[i] + gamma_od[i] - gamma_ds[i]
    
    gamma_los.append(gl)
    
    ##source model
                    
    e1 = random.uniform(source_ellipticity_min, source_ellipticity_max)
    e2 = random.uniform(source_ellipticity_min, source_ellipticity_max)
    source_ellipticity.append([e1,e2])

    source_amp.append(random.uniform(source_amp_min, source_amp_max))
    
    source_radius = random.uniform(R_min,R_max)
    
    #conversion between source radius in pc and source radius in arcsec
    r_sers = (source_radius/lensCosmo.ds)*(206265/10**6)
    
    R_sersic.append(r_sers)

    n_sersic.append(random.uniform(n_sersic_min, n_sersic_max))
    
    ##point source model
    
    amp_factor.append(random.uniform(amp_factor_min,amp_factor_max)*source_amp[i])
```

## Setting up the lens models

```python
# Generate the coordinate grid and image properties
kwargs_data = []
data_class = []

kwargs_data_E = []
kwargs_data_Q = []
kwargs_data_EQ = []

data_class_E = []
data_class_Q = []
data_class_EQ = []

# Generate the PSF variables
kwargs_psf = []
psf_class = []

# the lens model

lens_mock_list = []

# Parameters of the main lens
# We define its centre as the line of sight
ellipticity_lens = []
kwargs_epl = []

# LOS shears

kwargs_los = []

kwargs_lens = []
lens_mock_class = []

e_sersic = []
kwargs_sersic = []

kwargs_source = []

ra_source = []
dec_source = []

# for this example, we ignore the presence of deflector light
lens_light_model_list = []
kwargs_lens_light_list = []

# perturb observed magnification due to e.g. micro-lensing
mag_pert = []

x_image = []
y_image = []

kwargs_ps = [] 


imageModel_E = []    
image_sim_E = []
kwargs_model_E = []

imageModel_Q = []    
image_sim_Q = []
kwargs_model_Q = []

imageModel_EQ = []    
image_sim_EQ = []
kwargs_model_EQ = []

one_image_indices = []

for i in range(n_models):
    # Generate the coordinate grid and image properties
    kwargs_data.append(sim_util.data_configure_simple(numPix[i], deltaPix, exp_time, background_rms[i]))
    data_class.append(ImageData(**kwargs_data[i]))

    kwargs_data_E.append(sim_util.data_configure_simple(numPix[i], deltaPix, exp_time, background_rms[i]))
    kwargs_data_Q.append(sim_util.data_configure_simple(numPix[i], deltaPix, exp_time, background_rms[i]))
    kwargs_data_EQ.append(sim_util.data_configure_simple(numPix[i], deltaPix, exp_time, background_rms[i]))

    data_class_E.append(ImageData(**kwargs_data_E[i]))
    data_class_Q.append(ImageData(**kwargs_data_Q[i]))
    data_class_EQ.append(ImageData(**kwargs_data_EQ[i]))
    
    # Generate the PSF variables
    kwargs_psf.append({'psf_type': 'GAUSSIAN', 'fwhm': fwhm, 'pixel_size': deltaPix, 'truncation': 3})
    psf_class.append(PSF(**kwargs_psf[i]))

    # the lens model

    lens_mock_list.append([dominant_model[i], 'LOS'])

    # Parameters of the main lens
    # We define its centre as the line of sight
    ellipticity_lens.append(lens_ellipticity[i])
    kwargs_epl.append({'theta_E': theta_E[i],
                   'gamma': gamma[i],
                   'center_x': 0,
                   'center_y': 0,
                   'e1': ellipticity_lens[i][0],
                   'e2': ellipticity_lens[i][1]})

    # LOS shears

    kwargs_los.append({
            'kappa_od': 0,
            'kappa_os': 0, 'kappa_ds': 0,
            'omega_od': 0,
            'omega_os': 0, 'omega_ds': 0,
            'gamma1_od': gamma_od[i][0], 'gamma2_od': gamma_od[i][1],
            'gamma1_os': gamma_os[i][0], 'gamma2_os': gamma_os[i][1],
            'gamma1_ds': gamma_ds[i][0], 'gamma2_ds': gamma_ds[i][1]})
        

    kwargs_lens.append([kwargs_epl[i], kwargs_los[i]])
    lens_mock_class.append(LensModel(lens_model_list=lens_mock_list[i], z_lens=z_lens[i], z_source=z_source[i], cosmo=cosmo))

    lensEquationSolver = LensEquationSolver(lens_mock_class[i])
    
    single_image = True
    
    count = 0
    
    while single_image:
        
        count += 1
        
        ra_s = random.uniform(source_pos_min, source_pos_max)
        dec_s = random.uniform(source_pos_min, source_pos_max)
        
        xx, yy = lensEquationSolver.findBrightImage(ra_s, dec_s, kwargs_lens[i], numImages=4,
                                                          min_distance=deltaPix, search_window=numPix[i] * deltaPix)
    
        if len(xx) > 1:
            single_image = False
         
        if count > 10:
            one_image_indices.append(i)
            single_image = False

    ra_source.append(ra_s)
    dec_source.append(dec_s)   
    
    x_image.append(xx)
    y_image.append(yy)

    # Sersic parameters in the initial simulation
    e_sersic.append(source_ellipticity[i])
    kwargs_sersic.append({'amp': source_amp[i],
                     'R_sersic': R_sersic[i],
                     'n_sersic': n_sersic[i],
                     'e1': e_sersic[i][0],
                     'e2': e_sersic[i][1],
                     'center_x': ra_source[i],
                     'center_y': dec_source[i]})

    kwargs_source.append([kwargs_sersic[i]])
    
    # compute lensing magnification at image positions
    mag = lens_mock_class[i].magnification(x_image[i], y_image[i], kwargs=kwargs_lens[i])
    mag = np.abs(mag)  # ignore the sign of the magnification

    # perturb observed magnification due to e.g. micro-lensing
    mag_pert = np.random.normal(mag, 0.01, len(mag))

    point_amp = mag * amp_factor[i]  # multiply by intrinsic quasar brightness (in counts/s)
    kwargs_ps.append([{'ra_image': x_image[i], 'dec_image': y_image[i],
                               'point_amp': point_amp}])  # quasar point source position in the lens plane and intrinsic brightness
        
    ## E
    
    imageModel_E.append(ImageModel(data_class_E[i], psf_class[i], lens_mock_class[i], source_model_class,
                               kwargs_numerics=kwargs_numerics))
    
    image_sim_E.append(imageModel_E[i].image(kwargs_lens[i], 
                                       kwargs_source[i])) 

    #adding noise
    poisson = image_util.add_poisson(image_sim_E[i], exp_time=exp_time)
    bkg = image_util.add_background(image_sim_E[i], sigma_bkd=background_rms[i])
    image_sim_E[i] = image_sim_E[i] + bkg + poisson

    kwargs_data_E[i]['image_data'] = image_sim_E[i]
    data_class_E[i].update_data(image_sim_E[i])

    kwargs_model_E.append({'lens_model_list': lens_mock_list[i], 
                     'source_light_model_list': source_model_list})

    ## Q

    imageModel_Q.append(ImageModel(data_class_Q[i], psf_class[i], lens_mock_class[i],
                                point_source_class=point_source_class, 
                                kwargs_numerics=kwargs_numerics))
    
    image_sim_Q.append(imageModel_Q[i].image(kwargs_lens[i], 
                                 kwargs_source=None,
                                 kwargs_lens_light=None,
                                 lens_light_add = False,
                                 kwargs_ps=kwargs_ps[i]))

    #adding noise
    poisson = image_util.add_poisson(image_sim_Q[i], exp_time=exp_time)
    bkg = image_util.add_background(image_sim_Q[i], sigma_bkd=background_rms[i])
    image_sim_Q[i] = image_sim_Q[i] + bkg + poisson

    kwargs_data_Q[i]['image_data'] = image_sim_Q[i]
    data_class_Q[i].update_data(image_sim_Q[i])

    kwargs_model_Q.append({'lens_model_list': lens_mock_list[i],
                'point_source_model_list': point_source_list})

    ## EQ

    imageModel_EQ.append(ImageModel(data_class_EQ[i], psf_class[i], lens_mock_class[i],
                                    source_model_class,
                                    point_source_class=point_source_class, 
                                    kwargs_numerics=kwargs_numerics))
    
    image_sim_EQ.append(imageModel_EQ[i].image(kwargs_lens[i], 
                                 kwargs_source[i],
                                 kwargs_lens_light=None,
                                 lens_light_add = False,
                                 kwargs_ps=kwargs_ps[i]))

    #adding noise
    poisson = image_util.add_poisson(image_sim_EQ[i], exp_time=exp_time)
    bkg = image_util.add_background(image_sim_EQ[i], sigma_bkd=background_rms[i])
    image_sim_EQ[i] = image_sim_EQ[i] + bkg + poisson

    kwargs_data_EQ[i]['image_data'] = image_sim_EQ[i]
    data_class_EQ[i].update_data(image_sim_EQ[i])

    kwargs_model_EQ.append({'lens_model_list': lens_mock_list[i], 
                     'source_light_model_list': source_model_list,
                'point_source_model_list': point_source_list})
```

## removing models with 1 or no point source images

```python
j = 0

for i in one_image_indices:
    
    ###cosmology

    del z_source[i-j]
    del z_lens[i-j]
    
    ##data

    del background_rms[i-j]

    ##lens model

    del dominant_model[i-j]

    del lens_ellipticity[i-j]
    del theta_E[i-j]
    del gamma[i-j]

    del gamma_od[i-j]
    del gamma_os[i-j]
    del gamma_ds[i-j]

    del gamma_los[i-j]

    ##source model

    del source_ellipticity[i-j]

    del ra_source[i-j]
    del dec_source[i-j]

    del source_amp[i-j]

    del R_sersic[i-j]

    del n_sersic[i-j]

    ##point source model

    del amp_factor[i-j]

    # Generate the coordinate grid and image properties
    del kwargs_data[i-j]
    del data_class[i-j]

    del kwargs_data_E[i-j]
    del kwargs_data_Q[i-j]
    del kwargs_data_EQ[i-j]

    del data_class_E[i-j]
    del data_class_Q[i-j]
    del data_class_EQ[i-j]

    # Generate the PSF variables
    del kwargs_psf[i-j]
    del psf_class[i-j]

    # the lens model

    del lens_mock_list[i-j]

    # Parameters of the main lens
    # We define its centre as the line of sight
    del ellipticity_lens[i-j]
    del kwargs_epl[i-j]

    # LOS shears

    del kwargs_los[i-j]

    del kwargs_lens[i-j]
    del lens_mock_class[i-j]

    del e_sersic[i-j]
    del kwargs_sersic[i-j]

    del kwargs_source[i-j]

    del x_image[i-j]
    del y_image[i-j]

    del kwargs_ps[i-j]


    del imageModel_E[i-j]    
    del image_sim_E[i-j]
    del kwargs_model_E[i-j]

    del imageModel_Q[i-j]   
    del image_sim_Q[i-j]
    del kwargs_model_Q[i-j]

    del imageModel_EQ[i-j]   
    del image_sim_EQ[i-j]
    del kwargs_model_EQ[i-j]
    
    j += 1
    
n_models_use = len(z_source)
```

## Plotting Simulations

```python
f, axes = plt.subplots(n_models_use, 3,  figsize=(15, (n_models_use)*5), sharex=False, sharey=False)
ax = axes


#Image

cmap_string = 'gray'
cmap = plt.get_cmap(cmap_string)
cmap.set_bad(color='k', alpha=1.)
cmap.set_under('k')

v_min = -5
v_max = 3

for i in range(n_models_use):
    im = ax[i,0].matshow(np.log10(image_sim_E[i]), origin='lower', vmin=v_min, vmax=v_max, cmap=cmap, extent=[0, 0.2, 0, 0.2])
    im = ax[i,1].matshow(np.log10(image_sim_Q[i]), origin='lower', vmin=v_min, vmax=v_max, cmap=cmap, extent=[0, 0.2, 0, 0.2])
    im = ax[i,2].matshow(np.log10(image_sim_EQ[i]), origin='lower', vmin=v_min, vmax=v_max, cmap=cmap, extent=[0, 0.2, 0, 0.2])

plt.show()
```

```python
td_cosmo = []
t_days = []
dt_days = []
dt_sigma = []
dt_measured = []
ddt = []

for i in range(n_models_use):
    td_cosmo.append(TDCosmography(z_lens[i], z_source[i], kwargs_model_Q[i], cosmo_fiducial=cosmo))

    # time delays, the unit [days] is matched when the lensing angles are in arcsec
    t_days.append(td_cosmo[i].time_delays(kwargs_lens[i], kwargs_ps[i], kappa_ext=0))

    # relative delays (observable). The convention is relative to the first image
    dt_days.append(t_days[i][1:] - t_days[i][0])
    
    # and errors can be assigned to the measured relative delays (full covariance matrix not yet implemented)
    dt_sigmas = []
    
    #here, I want a random percentage error, and I want to apply this to each of the time delays
    for j in range(len(dt_days[i])):
        dt_sigmas.append(dt_days[i][j]*random.uniform(dt_sigma_factor_min, dt_sigma_factor_max))
        
    dt_sigma.append(dt_sigmas)
    
    # and here a realisation of the measurement with the quoted error bars
    dt_measured.append(np.random.normal(dt_days[i], dt_sigmas))

    #the time delay distance is
    lensCosmo = LensCosmo(z_lens=z_lens[i], z_source=z_source[i], cosmo=cosmo)
    ddt.append(lensCosmo.ddt)
```

##  Setting Up the Modelling

```python
num_source_model = len(source_model_list)

lens_model_list = []

kwargs_model_E = []
kwargs_model_Q = []
kwargs_model_EQ = []

kwargs_constraints_E = []
kwargs_constraints_Q = []
kwargs_constraints_EQ = []
kwargs_constraints_QT = []
kwargs_constraints_EQT = []


kwargs_likelihood_NTD = {'check_bounds': True,
                        'force_no_add_image': False,
                        'source_marg': False,
                        'image_position_uncertainty': 0.004,
                        'check_matched_source_position': True,
                        'source_position_tolerance': 0.001,
                        'source_position_sigma': 0.001,
                        'time_delay_likelihood': False,
                        }

kwargs_likelihood_TD = {'check_bounds': True,
                        'force_no_add_image': False,
                        'source_marg': False,
                        'image_position_uncertainty': 0.004,
                        'check_matched_source_position': True,
                        'source_position_tolerance': 0.001,
                        'source_position_sigma': 0.001,
                        'time_delay_likelihood': True,
                        }

multi_band_list_E = []
multi_band_list_Q = []
multi_band_list_EQ = []

kwargs_data_joint_E = []
kwargs_data_joint_Q = []
kwargs_data_joint_EQ = []
kwargs_data_joint_QT = []
kwargs_data_joint_EQT = []

#the parameters for the different models
                       
kwargs_params_E = []
kwargs_params_Q = []
kwargs_params_EQ = []
kwargs_params_QT = []
kwargs_params_EQT = []

gamma_SPT = []
e_eff_sersic = []
x_eff_sersic = []
y_eff_sersic = []

for i in range(n_models_use):

    lens_model_list.append(['EPL','LOS_MINIMAL'])

    kwargs_model_E.append({'lens_model_list': lens_model_list[i],
                'source_light_model_list': source_model_list})

    kwargs_model_Q.append({'lens_model_list': lens_model_list[i],
                    'point_source_model_list': point_source_list})
                          
    kwargs_model_EQ.append({'lens_model_list': lens_model_list[i],
                    'source_light_model_list': source_model_list,
                    'point_source_model_list': point_source_list})

                           
    kwargs_constraints_E.append({})

    kwargs_constraints_Q.append({'num_point_source_list': [len(x_image[i])], 
                          'Ddt_sampling': False,
                                  })
                                
    kwargs_constraints_EQ.append({'joint_source_with_point_source': [[0, 0]],
                          'num_point_source_list': [len(x_image[i])], 
                          'Ddt_sampling': False,
                                  })


    kwargs_constraints_QT.append({'num_point_source_list': [len(x_image[i])], 
                          'Ddt_sampling': True,
                                  })

    kwargs_constraints_EQT.append({'joint_source_with_point_source': [[0, 0]],
                          'num_point_source_list': [len(x_image[i])], 
                          'Ddt_sampling': True,
                                  })

    multi_band_list_E.append([[kwargs_data_E[i], kwargs_psf[i], kwargs_numerics]])
    multi_band_list_Q.append([[kwargs_data_Q[i], kwargs_psf[i], kwargs_numerics]])
    multi_band_list_EQ.append([[kwargs_data_EQ[i], kwargs_psf[i], kwargs_numerics]])

    kwargs_data_joint_E.append({'multi_band_list': multi_band_list_E[i],
                         'multi_band_type': 'multi-linear'})

    kwargs_data_joint_Q.append({'multi_band_list': multi_band_list_Q[i],
                         'multi_band_type': 'multi-linear'})

    kwargs_data_joint_EQ.append({'multi_band_list': multi_band_list_EQ[i], 
                             'multi_band_type': 'multi-linear',
                        'time_delays_measured': dt_measured[i],
                        'time_delays_uncertainties': dt_sigma[i]})

    kwargs_data_joint_QT.append({'multi_band_list': multi_band_list_Q[i],
                         'multi_band_type': 'multi-linear',
                        'time_delays_measured': dt_measured[i],
                        'time_delays_uncertainties': dt_sigma[i]})

    kwargs_data_joint_EQT.append({'multi_band_list': multi_band_list_EQ[i], 
                             'multi_band_type': 'multi-linear',
                        'time_delays_measured': dt_measured[i],
                        'time_delays_uncertainties': dt_sigma[i]})
      
    # Lens model

    fixed_lens = []
    kwargs_lens_init = []
    kwargs_lens_sigma = []
    kwargs_lower_lens = []
    kwargs_upper_lens = []

    # Parameters for the main-lens model
    # e_eff = e 
    fixed_lens.append({'center_x': 0, 'center_y': 0}) # fix the line of sight
    kwargs_lens_init.append({'theta_E': theta_E[i], 'gamma': gamma[i],
                             'e1': ellipticity_lens[i][0],
                             'e2': ellipticity_lens[i][1]})
    kwargs_lens_sigma.append({'theta_E': .0001, 'gamma': 0.1,
                              'e1': 0.01, 'e2': 0.01})
    kwargs_lower_lens.append({'theta_E': theta_E[i]*0.5, 'gamma': gamma[i]*0.5,
                              'e1': -0.7, 'e2': -0.7})
    kwargs_upper_lens.append({'theta_E': theta_E[i]*1.5,  'gamma': gamma[i]*1.5,
                              'e1': 0.7, 'e2': 0.7})

    # LOS corrections
    #gamma_los = gamma_os + gamma_od - gamma_ds
    fixed_lens.append({'kappa_od': 0, 'omega_od': 0, 'kappa_los': 0})
    los_keys = ['gamma1_od', 'gamma2_od', 'gamma1_los', 'gamma2_los', 'omega_los']
    kwargs_lens_init.append({'gamma1_od': gamma_od[i][0],
                             'gamma2_od': gamma_od[i][1],
                             'gamma1_los': gamma_los[i][0],
                             'gamma2_los': gamma_los[i][1],
                             'omega_los':0})
    kwargs_lens_sigma.append({key: 0.01 for key in los_keys})
    kwargs_lower_lens.append({key: -0.5 for key in los_keys})
    kwargs_upper_lens.append({key: 0.5 for key in los_keys})

    lens_params = [kwargs_lens_init,
                   kwargs_lens_sigma,
                   fixed_lens,
                   kwargs_lower_lens,
                   kwargs_upper_lens]

    # Source models

    fixed_source = []
    kwargs_source_init = []
    kwargs_source_sigma = []
    kwargs_lower_source = []
    kwargs_upper_source = []

    # adapt the initialisation of the model
    gamma_SPT.append(gamma_od[i] - gamma_ds[i])
    e_eff_sersic.append(e_sersic[i] - gamma_SPT[i])
    x_eff_sersic.append((1 - gamma_SPT[i][0]) * ra_source[i] - gamma_SPT[i][1] * dec_source[i])
    y_eff_sersic.append(- gamma_SPT[i][1] * ra_source[i] + (1 + gamma_SPT[i][0]) * dec_source[i])

    fixed_source.append({})
    kwargs_source_init.append({'amp': source_amp[i],
                                'R_sersic': R_sersic[i],
                                'n_sersic': n_sersic[i],
                                'e1': e_eff_sersic[i][0],
                                'e2': e_eff_sersic[i][1],
                                'center_x': x_eff_sersic[i],
                                'center_y': y_eff_sersic[i]})
    kwargs_source_sigma.append({'n_sersic': 0.001, 'R_sersic': 0.001,
                                'e1': 0.01, 'e2': 0.01,
                                'center_x': 0.0001, 'center_y': 0.0001})
    kwargs_lower_source.append({'e1': -0.7, 'e2': -0.7,
                                'R_sersic': R_sersic[i]*0.5, 'n_sersic': n_sersic[i]*0.5,
                                'center_x': -10, 'center_y': -10})
    kwargs_upper_source.append({'e1': 0.7, 'e2': 0.7,
                                'R_sersic': R_sersic[i]*1.5, 'n_sersic': n_sersic[i]*1.5,
                                'center_x': 10, 'center_y': 10})

    source_params = [kwargs_source_init, kwargs_source_sigma,
                     fixed_source, kwargs_lower_source, kwargs_upper_source]


    # Point source models

    ps_model_list = point_source_list

    fixed_ps = []
    kwargs_ps_init = []
    kwargs_ps_sigma = []
    kwargs_lower_ps = []
    kwargs_upper_ps = []

    fixed_ps.append({})
    kwargs_ps_init.append({'ra_image': x_image[i], 'dec_image': y_image[i]})
    kwargs_ps_sigma.append({'ra_image': [0.01] * len(x_image[i]), 'dec_image': [0.01] * len(x_image[i])})
    kwargs_lower_ps.append({'ra_image': -10 * np.ones_like(x_image[i]), 'dec_image': -10 * np.ones_like(y_image[i])})
    kwargs_upper_ps.append({'ra_image': 10 * np.ones_like(x_image[i]), 'dec_image': 10 * np.ones_like(y_image[i])})

    ps_params = [kwargs_ps_init, kwargs_ps_sigma,
                     fixed_ps, kwargs_lower_ps, kwargs_upper_ps]

    #Cosmo models

    cosmo_model_list = point_source_list

    fixed_cosmo = []
    kwargs_cosmo_init = {'D_dt': ddt[i]}
    kwargs_cosmo_sigma = {'D_dt': 50}
    kwargs_lower_cosmo = {'D_dt': ddt[i]*0.2}
    kwargs_upper_cosmo = {'D_dt': ddt[i]*3}

    cosmo_params = [kwargs_cosmo_init, kwargs_cosmo_sigma,
                     fixed_cosmo, kwargs_lower_cosmo, kwargs_upper_cosmo]


    #the parameters for the different models

    kwargs_params_E.append({'lens_model': lens_params,
                    'source_model': source_params})
    
    kwargs_params_Q.append({'lens_model': lens_params,
                    'point_source_model': ps_params})

    kwargs_params_EQ.append({'lens_model': lens_params,
                    'source_model': source_params,
                    'point_source_model': ps_params})

    kwargs_params_QT.append({'lens_model': lens_params,
                    'point_source_model': ps_params,
                    'special': cosmo_params})

    kwargs_params_EQT.append({'lens_model': lens_params,
                    'source_model': source_params,
                    'point_source_model': ps_params,
                    'special': cosmo_params})
```

## The Fitting

```python
fitting_seq_E = []
fitting_seq_Q = []
fitting_seq_EQ = []
fitting_seq_QT = []
fitting_seq_EQT = []


walker_ratio = 10
fitting_kwargs_list = [['MCMC',
                        {'n_burn': 1, 'n_run': 1,
                         'walkerRatio': walker_ratio, 'sigma_scale': 1., 'threadCount': 1}]]

chain_list_E = []
kwargs_result_E = []

chain_list_Q = []
kwargs_result_Q = []

chain_list_EQ = []
kwargs_result_EQ = []

chain_list_QT = []
kwargs_result_QT = []

chain_list_EQT = []
kwargs_result_EQT = []


for i in range(n_models_use):
    
    print('')
    print('')
    print(i+1,'/', n_models_use)    
    print('')
    print('')
    
    fitting_seq_E.append(FittingSequence(kwargs_data_joint_E[i], kwargs_model_E[i], kwargs_constraints_E[i], 
                                  kwargs_likelihood_NTD, kwargs_params_E[i]))

    fitting_seq_Q.append(FittingSequence(kwargs_data_joint_Q[i], kwargs_model_Q[i], kwargs_constraints_Q[i], 
                                  kwargs_likelihood_NTD, kwargs_params_Q[i]))

    fitting_seq_EQ.append(FittingSequence(kwargs_data_joint_EQ[i], kwargs_model_EQ[i], kwargs_constraints_EQ[i], 
                                          kwargs_likelihood_NTD, kwargs_params_EQ[i]))

    fitting_seq_QT.append(FittingSequence(kwargs_data_joint_QT[i], kwargs_model_Q[i], kwargs_constraints_QT[i], 
                                  kwargs_likelihood_TD, kwargs_params_QT[i]))

    fitting_seq_EQT.append(FittingSequence(kwargs_data_joint_EQT[i], kwargs_model_EQ[i], kwargs_constraints_EQT[i], 
                                          kwargs_likelihood_TD, kwargs_params_EQT[i]))
 
    
    chain_list_E.append(fitting_seq_E[i].fit_sequence(fitting_kwargs_list))
    kwargs_result_E.append(fitting_seq_E[i].best_fit())

    chain_list_Q.append(fitting_seq_Q[i].fit_sequence(fitting_kwargs_list))
    kwargs_result_Q.append(fitting_seq_Q[i].best_fit())
                           
    chain_list_EQ.append(fitting_seq_EQ[i].fit_sequence(fitting_kwargs_list))
    kwargs_result_EQ.append(fitting_seq_EQ[i].best_fit())

    chain_list_QT.append(fitting_seq_QT[i].fit_sequence(fitting_kwargs_list))
    kwargs_result_QT.append(fitting_seq_QT[i].best_fit())   

    chain_list_EQT.append(fitting_seq_EQT[i].fit_sequence(fitting_kwargs_list))
    kwargs_result_EQT.append(fitting_seq_EQT[i].best_fit())
```
